```python
import cv2
import imutils
import random
from glob import glob
def augment_video_rotage(video_path, output_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Create a video writer object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, 25.0, (640, 480))
    rd=random.randrange(-20, 20)
    print(rd)
    # Loop through the video frames
    while True:
        # Read the current frame
        ret, frame = cap.read()
        if ret == True:
            # Apply your desired augmentation to the current frame
            augmented_frame= imutils.rotate(frame,rd)
            # Write the augmented frame to the output video file
            out.write(augmented_frame)

            # Display the augmented frame if desired
            cv2.imshow('Augmented Video', augmented_frame)

            # Wait for the 'q' key to be pressed to exit
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break

    # Release the video writer object and close the video file
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import random
def augment_video_zoom(input_video_path, output_video_path):

    cap = cv2.VideoCapture(input_video_path)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    frame_width = 640
    frame_height = 480
    out = cv2.VideoWriter(output_video_path, fourcc, 25.0, (frame_width, frame_height))

    zoom_factor = random.uniform(1,1.5)
    #print(zoom_factor)
    while cap.isOpened():
        ret, frame = cap.read()
        if ret:
            zoomed_frame = cv2.resize(frame, (int(frame_width * zoom_factor), int(frame_height * zoom_factor)))
            x_offset = int((zoomed_frame.shape[1] - frame_width) / 2)
            y_offset = int((zoomed_frame.shape[0] - frame_height) / 2)
            cropped_frame = zoomed_frame[y_offset:y_offset+frame_height, x_offset:x_offset+frame_width]
            out.write(cropped_frame)
        else:
            break

    cap.release()
    out.release()
```

```python
import cv2
import imutils
import random
from glob import glob
def augment_video_mix(video_path, output_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    frame_width = 640
    frame_height = 480
    # Create a video writer object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, 25.0, (640, 480))
    zoom_factor = random.uniform(1,1.5)
    rd=random.randrange(-20, 20)
    # Loop through the video frames
    while True:
        # Read the current frame
        ret, frame = cap.read()
        if ret == True:
            # Apply your desired augmentation to the current frame
            zoomed_frame = cv2.resize(frame, (int(frame_width * zoom_factor), int(frame_height * zoom_factor)))
            x_offset = int((zoomed_frame.shape[1] - frame_width) / 2)
            y_offset = int((zoomed_frame.shape[0] - frame_height) / 2)
            cropped_frame = zoomed_frame[y_offset:y_offset+frame_height, x_offset:x_offset+frame_width]
            
            augmented_frame= imutils.rotate(cropped_frame,rd)
            # Write the augmented frame to the output video file
            out.write(augmented_frame)

            # Display the augmented frame if desired
            cv2.imshow('Augmented Video', augmented_frame)

            # Wait for the 'q' key to be pressed to exit
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break

    # Release the video writer object and close the video file
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```
